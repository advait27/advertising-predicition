```python
#prediction of sales
#understanding the train and test data
#finding ut the shape of the data
print('\nShape of the training data :',ad.shape)

train_x = ad.drop(columns=['sales'], axis=1)
train_y = ad['sales']
```

```python
model = LinearRegression()
```

```python
model.fit(train_x,train_y)

#coefficients of the train model
print('\nCoefficient of model: ', model.coef_)

#intercept of the model
print('\nInterncept of model: ', model.intercept_)

predict = model.predict(train_x)
print('\nsales on training data', predict)

#rmse on training dataset
rmse_train = mean_squared_error(train_y, predict)**(0.5)
print('\nRMSE on train dataset: ', rmse_train)

predict_more = model.predict(train_x)
print('\nsales on data', predict_more)

```
